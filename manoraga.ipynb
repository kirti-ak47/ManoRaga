```python
import google.generativeai as genai
import os
from getpass import getpass

os.environ["GEMINI_API_KEY"] = getpass("Enter your Gemini API Key: ")
genai.configure(api_key=os.environ["GEMINI_API_KEY"])

```

```python
model = genai.GenerativeModel("models/gemini-flash-latest")
```

```python
import pandas as pd

csv_url = "https://raw.githubusercontent.com/kirti-ak47/ManoRaga/main/raga_data.csv"

raga_df = pd.read_csv(csv_url)

```

```python
def generate_raga_recommendation(user_feeling):
    raga_info = raga_df.to_string(index=False)

    prompt = f"""
You are a neuromusicology expert and Indian classical music scholar.

Below is a dataset of Indian ragas and their emotional benefits:
{raga_info}

User emotional state:
"{user_feeling}"

Tasks:
1. Identify the dominant emotion.
2. Choose the BEST matching raga ONLY from the dataset.
3. Explain how this raga helps mental and emotional well-being.
4. Keep the explanation calming, simple, and supportive.
"""

    response = model.generate_content(prompt)
    return response.text

```

```python
user_input = input("Describe how you are feeling right now: ")
```

```python
result = generate_raga_recommendation(user_input)
print("\n ManoRaga Recommendation\n")
print(result)
```
